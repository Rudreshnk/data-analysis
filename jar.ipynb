```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Walmart Sales dataset
# Assume the dataset is named 'walmart_sales.xlsx' and is in the same directory as the script
data = pd.read_excel(r'C:\Users\rudre\Walmart_Sales.xlsx')
print("A.Analyze the performance of sales and revenue at the city and branch level")
# Group by City and calculate total sales and revenue
city_sales_revenue = data.groupby('City').agg({'Unit price': 'sum', 'Quantity': 'sum'}).reset_index()
city_sales_revenue['Revenue'] = city_sales_revenue['Unit price'] * city_sales_revenue['Quantity']

# Group by Branch and calculate total sales and revenue
branch_sales_revenue = data.groupby('Branch').agg({'Unit price': 'sum', 'Quantity': 'sum'}).reset_index()
branch_sales_revenue['Revenue'] = branch_sales_revenue['Unit price'] * branch_sales_revenue['Quantity']

# Print the results
print("Sales and Revenue by City:")
print(city_sales_revenue)

print("\nSales and Revenue by Branch:")
print(branch_sales_revenue)

# Plotting
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='City', y='Quantity', data=city_sales_revenue)
plt.title('Total Sales Quantity by City')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Quantity')

plt.subplot(1, 2, 2)
sns.barplot(x='City', y='Revenue', data=city_sales_revenue)
plt.title('Total Revenue by City')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Revenue')

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Quantity', data=branch_sales_revenue)
plt.title('Total Sales Quantity by Branch')
plt.xticks(rotation=45)
plt.xlabel('Branch')
plt.ylabel('Quantity')

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Revenue', data=branch_sales_revenue)
plt.title('Total Revenue by Branch')
plt.xticks(rotation=45)
plt.xlabel('Branch')
plt.ylabel('Revenue')

plt.tight_layout()
plt.show()

```

```python
# Group by Branch and City and calculate the average price
avg_price_by_branch_city = data.groupby(['Branch', 'City']).agg({'Unit price': 'mean'}).reset_index()

# Display the average price of an item sold at each branch of the city
print("Average price of an item sold at each branch of the city:")
print(avg_price_by_branch_city)
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Print unique values in the 'Date' column to check the date format
print(data['Date'].unique())

# Extract month and year from the 'Date' column
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Print unique values of 'Month' and 'Year' to verify data for April 2019 exists
print(data['Month'].unique())
print(data['Year'].unique())

# Filter the data for April 2019
april_2019_data = data[(data['Month'] == 4) & (data['Year'] == 2019)]

# Group by product line, gender, and payment method and calculate total sales and revenue
sales_revenue_analysis = april_2019_data.groupby(['Product line', 'Gender', 'Payment']).agg({'Unit price': 'sum', 'Quantity': 'sum'}).reset_index()
sales_revenue_analysis['Revenue'] = sales_revenue_analysis['Unit price'] * sales_revenue_analysis['Quantity']

# Print the analysis results
print("Performance of sales and revenue for April 2019, by Product line, Gender, and Payment Method:")
print(sales_revenue_analysis)
```
